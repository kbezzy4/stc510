```python
# Importing the needed modules
import pandas as pd
import numpy as np
import datetime as dt
```

```python
# Pulling in the data
data = pd.read_csv('crimedata.csv')
```

```python
# Making sense of the data by getting info on columns:
df = pd.DataFrame(data)
df.info()
```

```python
# Renaming columns because of spacing in order to reference easier and confirming the renaming:
df.rename(columns={'INC NUMBER': 'INC_NUMBER'}, inplace=True)
df.rename(columns={'OCCURRED ON': 'OCCURRED_ON'}, inplace=True)
df.rename(columns={'OCCURRED TO': 'OCCURRED_TO'}, inplace=True)
df.rename(columns={'UCR CRIME CATEGORY': 'UCR_CRIME_CATEGORY'}, inplace=True)
df.rename(columns={'100 BLOCK ADDR': '100_BLOCK_ADDR'}, inplace=True)
df.rename(columns={'PREMISE TYPE': 'PREMISE_TYPE'}, inplace=True)
df
```

```python
# Getting some value counts for the crime category
df.UCR_CRIME_CATEGORY.value_counts()
```

```python
# To find crime rate in 2023, only have crimes in 2023; converting "occurred on" to datetime and filtering
df['OCCURRED_ON'] = pd.to_datetime(df['OCCURRED_ON'])
res = df.loc[(df['OCCURRED_ON'] >= '2023-01-01') & (df['OCCURRED_ON']<='2024-01-01')]
print(res)
```

```python
# Getting value counts for crimes that year
res.UCR_CRIME_CATEGORY.value_counts()
```

```python
# Population data needed for crime rate; population in 2023: 7,431,344
pop_az_23 = 7431344
x1 = np.array([25705,8513,6518,6162,5213,1807,884,693,187])
crime_rate_23 = ((x1/pop_az_23)*100000)
crime_rate_23
```

```python
#Repeating last three lines of code, but for year 2018. AZ pop. in 2018: 7,164,000
df['OCCURRED_ON'] = pd.to_datetime(df['OCCURRED_ON'])
res2 = df.loc[(df['OCCURRED_ON'] >= '2018-01-01') & (df['OCCURRED_ON']<='2019-01-01')]
print(res2)
```

```python
res2.UCR_CRIME_CATEGORY.value_counts()
```

```python
pop_az_18 = 7164000
x2 = np.array([33203,9303,6910,5734,5393,2820,1049,460,113])
crime_rate_18 = ((x2/pop_az_18)*100000)
crime_rate_18
```

```python
# Getting percent change in crime rate from 2018 to 2023
crime_rate_diff = ((crime_rate_23)-(crime_rate_18)/(abs(crime_rate_18))*100)
crime_rate_diff
```



---


LARCENY-THEFT had the greatest percent change from 2018 - 2023 while MURDER and NON-NEGLIGENT MANSLAUGHTER had the lowest percent change from 2018 - 2023

```python
# Sort the dataframe by crime category and zip code
df.sort_values(by=['UCR_CRIME_CATEGORY','ZIP'])
```

```python
#Relative Frequencies for different crimes using lambda function
df.groupby('UCR_CRIME_CATEGORY').apply(lambda x: x['UCR_CRIME_CATEGORY'].count()/df.shape[0])
```
